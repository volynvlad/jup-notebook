```python
#convert categorical variable into dummy
df_train = pd.get_dummies(df_train)
```

```python
X = df_train.drop(['SalePrice', 'Id'], axis=1)
y = df_train.SalePrice
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=.33)
```

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
```

```python
model = lr.fit(X_train, y_train)
```

```python
print ("R^2 is: \n", model.score(X_test, y_test))
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print ('RMSE is: \n', mean_squared_error(y_test, predictions))
```

```python
actual_values = y_test
plt.scatter(predictions, actual_values, alpha=.7, color='b') #alpha helps to show overlapping data
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Linear Regression Model')
plt.show()
```

```python
best_alhpa = 3**(-4)
mean = 1
for i in range (-4, 5):
    alpha = 3**i
    rm = linear_model.Ridge(alpha=alpha)
    ridge_model = rm.fit(X_train, y_train)
    preds_ridge = ridge_model.predict(X_test)

    plt.scatter(preds_ridge, actual_values, alpha=.75, color='b')
    plt.xlabel('Predicted Price')
    plt.ylabel('Actual Price')
    plt.title('Ridge Regularization with alpha = {}'.format(alpha))
    overlay = 'R^2 is: {}\nRMSE is: {}'.format(
                    ridge_model.score(X_test, y_test),
                    mean_squared_error(y_test, preds_ridge))
    if mean > mean_squared_error(y_test, preds_ridge):
        best_alhpa = alpha
        mean = mean_squared_error(y_test, preds_ridge)
    plt.annotate(s=overlay, xy=(12.1, 10.6), size='x-large')
    plt.show()
print('best alpha is: {} | rmse is: {}'.format(best_alhpa, mean))
```
